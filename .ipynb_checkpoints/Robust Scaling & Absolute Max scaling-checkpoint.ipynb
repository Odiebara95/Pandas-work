## absolute maximum scaling

```python
# in this scaling, we divide the column we want to scale with the absolute max value of that column.
maxConf=usa.Confirmed.abs().max()
(usa.Confirmed)/maxConf
```

```python
usa["Confirmed_scaled"]=(usa.Confirmed)/maxConf
```

```python
usa.head()
```

## Robust Scaling

```python
usa.describe()
```

```python
# Robust scaling transforms the feature vector by subtracting the median and then
# dividing by the interquartile range (75% value - 25% value).
```

```python
median=usa.Deaths.median()
Deaths_75 = 418634.5
Deaths_25 = 3935

# robust sclaing formula

robustScaling=((usa.Deaths-median)/(Deaths_75 - Deaths_25))
usa["Deaths_scaled"]=robustScaling
```
