## missing values in column FIPS & Admin2
## we take care of them using backward fill

```python
df1=df.fillna(method="bfill",)
```

```python
df_modified = df1.fillna(method="ffill")
```

```python
df_modified.iso2.unique()
```

```python
df_modified.iso3.unique()
```

```python
df_modified["iso2"]=df_modified.iso2.map({"AS":1,"GU":2,"MP":3,"PR":4,"VI":5,"US":6})
df_modified["iso3"]=df_modified.iso3.map({"ASM":1,"GUM":2,"MNP":3,"PRI":4,"VIR":5,"USA":6})
```

```python
df_modified
```

```python
df_modified.memory_usage(deep=True).sum()
```

```python
df_modified["iso2"]=df_modified.iso2.astype("int8")
df_modified["iso3"]=df_modified.iso3.astype("int8")
```

```python
df_modified.memory_usage(deep=True).sum()
```

```python
df_modified.memory_usage().sum()
```

```python
df_modified.info()
```

```python
df_modified["Confirmed"]=df_modified.Confirmed.astype("int16")
df_modified["Deaths"]=df_modified.Deaths.astype("int16")
```

```python
df_modified.memory_usage()
```

```python
df_final=df_modified
```

```python
df_final["Country_Region"]=df_final.Country_Region.astype("category")
```

```python
df_final.memory_usage().sum()
```

```python
df_final.memory_usage(deep=True).sum()
```
