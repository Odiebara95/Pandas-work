```python
import pandas as pd

details = pd.DataFrame({'First Name': ['Paul','Sophia','George','Kindness','Michael'],
                       'Last Name': ['Odiebara','Jackrich','Smith','Silver','Madu'],
                       'Type': ['Full-time Employee','Intern','Full-time Employee','Part-time Employee','Full-time Employee'],
                       'Department': ['Administration','Technical','Administration','Technical','Management'],
                       'Year(s)_of_Employment': [4,2,5,7,6],
                       'Salary': [15000,5000,10000,20000,15000]
                       })
details
```

```python
pd.pivot_table(data=details,index=['Type'],columns=['Department'],values='Salary',aggfunc='mean')
```

```python
# pivot tables with multiple arggfuncs
pd.pivot_table(data=details,index=['Type'],
               values='Salary',aggfunc=['sum','mean','count'],
              margins=True,margins_name='Grand Total')
```

## DataFrames Concatenation

```python
import numpy as np
```

```python
smartHub = {
    "Classes": ['Data Science', 'App Dev', 'Web Dev'],
    "Time": ['9-11', '11-2', '2-5']
}

itClass= pd.DataFrame(smartHub, index = 
                 ["Mon", "Wed", "Fri"])
itClass
```

```python
vertical_concat = pd.concat([details, itClass], axis=0, sort=True)
vertical_concat
```

```python
Horizontal_concat = pd.concat([details, itClass], axis=1, sort=False)
Horizontal_concat
```
